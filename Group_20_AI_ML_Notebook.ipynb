```python
# ============================================
# STEP 4: INITIAL DATA PREPARATION
# ============================================
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')
```

```python
# ---- 4.1 Fix Invalid Categories  ----
df = pd.read_csv('UCI_Credit_Card.csv')
df_clean = df.copy()

df_clean['EDUCATION'] = df_clean['EDUCATION'].replace({0:4, 5:4, 6:4})
df_clean['MARRIAGE'] = df_clean['MARRIAGE'].replace({0:3})

print("✔️ Invalid categories successfully fixed!")
print("\nExample values after cleaning:")
print("EDUCATION (sample):", df_clean['EDUCATION'].head().tolist())
print("MARRIAGE (sample): ", df_clean['MARRIAGE'].head().tolist())
```

```python
# ---- 4.2 Encode Categorical Variables  ----
from sklearn.preprocessing import OrdinalEncoder

ordinal_cols = ['EDUCATION','MARRIAGE','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']
df_clean[ordinal_cols] = OrdinalEncoder().fit_transform(df_clean[ordinal_cols])


df_clean['SEX'] = df_clean['SEX'].replace({1:0, 2:1})

print("✔️ Categorical variables successfully encoded.")
print("\nExample encoded values:")
print(df_clean[['EDUCATION','MARRIAGE','SEX']].head())
```

```python
# ---- 4.3 Handle outliers (IQR capping)  ----
import numpy as np

def cap_outliers(df, col):
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.clip(df[col], lower, upper)


cols_to_cap = ['LIMIT_BAL'] + [f'BILL_AMT{i}' for i in range(1, 7)] + [f'PAY_AMT{i}' for i in range(1, 7)]

for col in cols_to_cap:
    cap_outliers(df_clean, col)


print("✔️ Outliers successfully capped using IQR method.")
print("\nSample of capped values (first 5 rows):")
print(df_clean[cols_to_cap].head())
```

```python
# ---- 4.4 Handle Class Imbalance with SMOTE and Scale Features ----

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# Features and target
X = df_clean.drop("default.payment.next.month", axis=1)
y = df_clean["default.payment.next.month"]

# ---- Handle class imbalance with SMOTE ----
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

# ---- Scale features ----
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_res)

# Convert back to DataFrame for readability
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)


print("✔️ SMOTE applied and features scaled successfully.")
print("\nClass distribution after SMOTE:")
print(pd.Series(y_res).value_counts())
print("\nSample of scaled features (first 5 rows):")
print(X_scaled_df.head())
```

```python
# ---- 4.5 K–Distribution (SelectKBest) to Score Features ----
from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

selector = SelectKBest(score_func=f_classif, k='all')  # k='all' to get scores for all features
selector.fit(X_scaled, y_res)

# DataFrame to view scores
feature_scores = pd.DataFrame({
    'Feature': X.columns,
    'Score': selector.scores_
}).sort_values(by='Score', ascending=False)

feature_scores
```

```python
# ---- 4.6 Dropping Low-Scoring Features ----
from sklearn.feature_selection import SelectKBest, f_classif
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Top 15 Features 
selector_top = SelectKBest(score_func=f_classif, k=15)
X_selected = selector_top.fit_transform(X_scaled, y_res)

selected_columns = X.columns[selector_top.get_support()]

feature_scores = pd.DataFrame({
    'Feature': X.columns,
    'Score': f_classif(X_scaled, y_res)[0]
})

# top 15 features
top_features_scores = feature_scores[feature_scores['Feature'].isin(selected_columns)].sort_values(by='Score', ascending=True)


plt.figure(figsize=(10,6))
sns.barplot(x='Score', y='Feature', data=top_features_scores, palette='magma')
plt.title('Top 15 Features Selected by SelectKBest')
plt.xlabel('ANOVA F-score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()

print("✔️ Selected top 15 features:", list(selected_columns))
```

```python
 # ---- 4.7 Spliting Data Into Train–Test Sets----
from sklearn.model_selection import train_test_split
import pandas as pd

X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y_res, test_size=0.2, random_state=42
)

print("Shapes after train-test split:")
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
# ============================================
# STEP 5: ITERATIVE MODELING
# ============================================
```

```python
# ---- 5.1 Logistic Regression (Baseline)----
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

lr = LogisticRegression(max_iter=2000)
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)

acc_lr = accuracy_score(y_test, pred_lr)
acc_lr
```

```python
# ---- 5.2 One Hot Coding----
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

cat_cols = ['SEX','EDUCATION','MARRIAGE','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6']
num_cols = [c for c in X.columns if c not in cat_cols]

preprocess = ColumnTransformer(
    [('num', StandardScaler(), num_cols),
     ('cat', OneHotEncoder(drop='first'), cat_cols)]
)

X_ohe = preprocess.fit_transform(X_res)

X_train_o, X_test_o, y_train_o, y_test_o = train_test_split(
    X_ohe, y_res, test_size=0.2, random_state=42
)

lr_o = LogisticRegression(max_iter=2000)
lr_o.fit(X_train_o, y_train_o)
pred_ohe = lr_o.predict(X_test_o)

acc_ohe = accuracy_score(y_test_o, pred_ohe)
acc_ohe
```

```python
# ---- 5.3 Threshold Tuning----
proba = lr.predict_proba(X_test)[:, 1]   # works now!!
pred_thresh = (proba >= 0.45).astype(int)

acc_thresh = accuracy_score(y_test, pred_thresh)
acc_thresh
```

```python
# ---- 5.4 Decision Tree ----
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt


dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)


dt_pred = dt.predict(X_test)


acc_dt = accuracy_score(y_test, dt_pred)
print("Decision Tree Accuracy:", acc_dt)


print(classification_report(y_test, dt_pred))


# ---- Ploting Decision Tree (top 3 levels) ----
plt.figure(figsize=(20, 10))
plot_tree(
    dt,
    feature_names=X.columns,
    class_names=['No Default', 'Default'],
    filled=True,
    max_depth=3
)
plt.show()
```

```python
# ---- 5.5 Random Forest Classifier----
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

rf = RandomForestClassifier(n_estimators=300, random_state=42)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

acc_rf = accuracy_score(y_test, pred_rf)
acc_rf
```

```python
# ---- 6.1 Accuracy Summary----
results = {
    "Logistic Regression": acc_lr,
    "Decision Tree": acc_dt,        
    "XGBoost": acc_xgb,
    "CatBoost": acc_cat,
    "Threshold LR": acc_thresh,
    "OHE Logistic Regression": acc_ohe
}

summary = pd.DataFrame(results, index=["Accuracy"]).T
print(summary)

# ---- 6.2 Best Model ----
best_model_name = summary['Accuracy'].idxmax()
best_accuracy = summary.loc[best_model_name, 'Accuracy']

best_model_name, best_accuracy
```
